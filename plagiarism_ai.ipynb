```python
import pandas as pd
import nltk
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "Machine learning is a branch of artificial intelligence.",
    "Artificial intelligence includes machine learning techniques.",
    "I love playing cricket in the evening.",
    "Machine learning is a branch of artificial intelligence."
]
```

```python
def preprocess(text):
    text = text.lower()

    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    return text
```

```python
processed_docs = [preprocess(doc) for doc in documents]

print(processed_docs)
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(processed_docs)
```

```python
print(tfidf_matrix.shape)
```

```python
similarity_matrix = cosine_similarity(tfidf_matrix)
print(similarity_matrix)
```

```python
threshold = 0.7
```

```python
n = len(documents)

for i in range(n):
    for j in range(i + 1, n):

        score = similarity_matrix[i][j]

        if score > threshold:
            print(f"Possible plagiarism between Document {i+1} and Document {j+1}")
            print(f"Similarity Score: {score:.2f}\n")
```

```python
n = len(documents)

for i in range(n):
    for j in range(i + 1, n):

        score = similarity_matrix[i][j]

        if score > threshold:
            print(f"Possible plagiarism between Document {i+1} and Document {j+1}")
            print(f"Similarity Score: {score:.2f}\n")
```

```python
results = []

for i in range(n):
    for j in range(i+1, n):

        score = similarity_matrix[i][j]

        if score > threshold:
            results.append([
                i+1,
                j+1,
                round(score*100,2)
            ])

report = pd.DataFrame(
    results,
    columns=[
        "Document A",
        "Document B",
        "Similarity (%)"
    ]
)

print(report)
report.to_csv("plagiarism_report.csv", index=False)
```
